```python
#Estrutura do número de telefone japonês: +81 XX-XXXX-XXXX ou +81 X-XXXX-XXXX

def is_jp_phone_number(input):
  if input[0:3] == "+81" and input[3] == " " and (input[4:6].isdecimal() or input[4].isdecimal()) and ((input[6] == "-" or input[6] == " ") or (input[5] == "-" or input[5] == " ")) and (input[7:11].isdecimal() or input[6:10].isdecimal()) and ((input[11] == "-" or input[11] == " ") or (input[10] == "-" or input[10] == " ")) and (input[12:16].isdecimal() or input[11:15].isdecimal()):
    print("É um número de telefone japonês\n")
  else:
    print("Não é um número de telefone japonês\n")

is_jp_phone_number("+81 99-9999-9999")
is_jp_phone_number("+81 9-9999-9999")
is_jp_phone_number("+81 88 8888 8888")
is_jp_phone_number("+81 8 8888 8888")
is_jp_phone_number("Japão")
```
