```python
#Estrutura do número de telefone fixo brasileiro: +55 XX XXXX-XXXX
#Estrutura do número de telefone móvel brasileiro: +55 XX XXXXX-XXXX

def is_br_phone_number(input):
    if input[0:3] == '+55' and input[3] == ' ' and input[4:6].isdecimal() and  input[6] == ' ' and input[7:12].isdecimal() and input[12] == '-' and input[13:17].isdecimal():
      print("É um número de telefone móvel brasileiro\n")
    elif input[0:3] == '+55' and input[3] == ' ' and input[4:6].isdecimal() and  input[6] == ' ' and  input[7:11].isdecimal() and input[11] == '-' and input[13:16].isdecimal():
      print("É um número de telefone fixo brasileiro\n")
    else:
      print("Não é um número de telefone brasileiro")

is_br_phone_number("+55 11 91014-6921")
is_br_phone_number("+55 11 9101-6921")
is_br_phone_number("Brasil")
```
