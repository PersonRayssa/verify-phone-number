```python
#Estrutura do número de telefone fixo chinês: +86 XX XXXX XXXX
#Estrutura do número de telefone móvel chinês: +86 XX XXXXX XXXX

def is_cn_phone_number(input):
    if input[0:3] == "+86" and input[3] == " " and input[4:6].isdecimal() and (input[6] == " " or input[6] == "-") and input[7:11].isdecimal() and (input[11] == " " or input[11] == "-") and input[12:16].isdecimal():
      print("É um número de telefone fixo chinês\n")
    elif input[0:3] == "+86" and input[3] == " " and input[4:6].isdecimal() and (input[6] == " " or input[6] == "-") and input[7:12].isdecimal() and (input[12] == " " or input[12] == "-") and input[13:17].isdecimal():
      print("É um número de telefone móvel chinês\n")
    else:
      print("Não é um número de telefone chinês")

is_cn_phone_number("+86 91 2345 9894")
is_cn_phone_number("+86 91-2345-9894")
is_cn_phone_number("+86 91 23451 9894")
is_cn_phone_number("+86 91-23451-9894")
is_cn_phone_number("China")
```
